# Cargar base de datos en formato accdb 
Una vez se carga la base de datos, las tablas contenidas son exportadas en formato .csv. Adicionalmente se imprime el nombre de todas las tablas contenidas en la base de datos
## Guia de uso
1. Establecer la direccion de la base de datos en 
- HOUSING_PATH = os.path.join("/home","gcpds","Descargas")

```python
HOUSING_PATH = os.path.join("/home","gcpds","Descargas")
db_access = os.path.join(HOUSING_PATH,"BD_PRUEBA.accdb")
for table in mdb.list_tables(db_access):
    print(table)
    df = mdb.read_table(db_access, table)
    df.to_csv(f'{table}.csv')
```
